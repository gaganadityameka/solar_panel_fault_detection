```python
import os
import shutil
import random
from PIL import Image

input_dir = "/kaggle/input/input1/Faulty_solar_panel/"
output_dir = "/kaggle/working/augmented_images_triple/"

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to perform image augmentation
def augment_images(class_dir, class_name):
    files = os.listdir(class_dir)
    for file in files:
        img_path = os.path.join(class_dir, file)
        try:
            img = Image.open(img_path)
            # Define augmentation operations (e.g., rotation, flipping, etc.)
            # Here, you can perform rotation by a random angle between -30 and 30 degrees
            angle = random.randint(-30, 30)
            rotated_img = img.rotate(angle)
            # You can add more augmentation operations as needed
            
            # Save the original image
            img.save(os.path.join(output_dir, class_name, file))
            # Save the augmented image
            rotated_img.save(os.path.join(output_dir, class_name, f"{file.split('.')[0]}_augmented.jpg"))
        except Exception as e:
            print(f"Error processing {img_path}: {e}")

# Perform augmentation for each class
for class_name in classes:
    class_dir = os.path.join(input_dir, class_name)
    if os.path.isdir(class_dir):
        # Create a subdirectory for each class in the output directory
        output_class_dir = os.path.join(output_dir, class_name)
        if not os.path.exists(output_class_dir):
            os.makedirs(output_class_dir)
        # Augment images for the current class
        augment_images(class_dir, class_name)

print("Augmentation completed.")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define the base directory containing the image directories
base_dir = "/kaggle/working/augmented_images_triple/"

# Define the directories for train, test, and validation
train_dir = "/kaggle/working/train/"
test_dir = "/kaggle/working/test/"
val_dir = "/kaggle/working/val/"

# Create directories for train, test, and validation sets
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Define the subdirectories containing the different classes of images
classes = ['Dusty', 'Clean', 'Snow-Covered', 'Electrical-damage', 'Physical-Damage', 'Bird-drop']

# Define the ratios for train, test, and validation sets
train_ratio = 0.7
test_ratio = 0.15
val_ratio = 0.15

# Function to split data and move images to the respective directories
def split_and_move_images(class_name):
    # Get the directory containing the images for the current class
    class_dir = os.path.join(base_dir, class_name)
    
    # Split the data into train, test, and validation sets
    train_files, test_val_files = train_test_split(os.listdir(class_dir), test_size=(1 - train_ratio), random_state=42)
    test_files, val_files = train_test_split(test_val_files, test_size=test_ratio / (test_ratio + val_ratio), random_state=42)
    
    # Move images to train directory
    for file in train_files:
        src = os.path.join(class_dir, file)
        dst = os.path.join(train_dir, class_name, file)
        os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
        shutil.copy(src, dst)
    
    # Move images to test directory
    for file in test_files:
        src = os.path.join(class_dir, file)
        dst = os.path.join(test_dir, class_name, file)
        os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)
        shutil.copy(src, dst)
    
    # Move images to validation directory
    for file in val_files:
        src = os.path.join(class_dir, file)
        dst = os.path.join(val_dir, class_name, file)
        os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
        shutil.copy(src, dst)

# Split and move images for each class
for class_name in classes:
    split_and_move_images(class_name)

# Function to count the number of images in a directory
def count_images(directory):
    count = sum([len(files) for r, d, files in os.walk(directory)])
    return count

# Count the number of images in train, test, and validation directories
train_count = count_images(train_dir)
test_count = count_images(test_dir)
val_count = count_images(val_dir)

print("Number of images in train directory:", train_count)
print("Number of images in test directory:", test_count)
print("Number of images in validation directory:", val_count)
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Define the directories for train, test, and validation
train_dir = "/kaggle/working/train/"
test_dir = "/kaggle/working/test/"
val_dir = "/kaggle/working/val/"

# Define image dimensions
img_height, img_width = 224, 224

# Create ImageDataGenerators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

# Prepare data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='categorical'
)






```

```python
# Load the InceptionV3 model
base_model = InceptionV3(weights='imagenet', include_top=False)
```

```python
# Add a global average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)

# Add a fully connected layer
x = Dense(1024, activation='relu')(x)

# Add a classification layer
predictions = Dense(len(train_generator.class_indices), activation='softmax')(x)
```

```python
# Create the model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)
```

```python
# Evaluate the model on test data
test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy:", test_acc)
```
